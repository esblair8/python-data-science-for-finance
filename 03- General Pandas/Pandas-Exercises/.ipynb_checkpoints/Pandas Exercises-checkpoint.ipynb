** What are the top 5 states with the most failed banks? **

```python
# CODE HERE
banklist.groupby('ST').count().sort_values('Bank Name', ascending=False).head()['Bank Name']
```

** What are the top 5 acquiring institutions? **

```python
# CODE HERE
banklist['Acquiring Institution'].value_counts()[:5]
```

** How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?**

```python
# CODE HERE
banklist[banklist['Acquiring Institution'] == 'State Bank of Texas']
```

** What is the most common city in California for a bank to fail in?**

```python
# CODE HERE
banklist[banklist['ST'] == 'CA'].groupby('City').count().sort_values('Bank Name', ascending=False)[:1]
```

** How many failed banks don't have the word "Bank" in their name? **

```python
# CODE HERE
sum(banklist['Bank Name'].apply(lambda name: 'Bank' not in name))
```

** How many bank names start with the letter 's' ? **

```python
# CODE HERE
banklist['Bank Name'].apply(lambda name: name[0].upper == 'S')
```
